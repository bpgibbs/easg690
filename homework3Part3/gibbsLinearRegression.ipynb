The following data comes from the paper Uptake and Fractionation of Thallium by Brassica juncea in a
Geogenic Thallium-Amended Substrate

I am doing a linear regression on ε205Tl values (an isotope fractionation value) from stems to seed pods of amazonite derived ε205Tl values vs hendricksite ε205Tl.

For a little bit of backgrounds, ε205Tl are representative of the ratio of 205Tl to 203T. Higher levels of ε205Tl are belived to be associated with the much more toxic protonated variant of thallium, so Shelby's lab is working to understand the ε205Tl values as plants uptake the metal from various substrates. Do substrates with different thallium sources result in different ε205Tl values throughout plants? This linear regression should provide insight into the correlation of amazonite and hendricksite substrate ε205Tl values. 

I used AI to assist in the creation of the following notebook.

```python
#import libraires

import numpy as np
import matplotlib.pyplot as plt
import scipy
import pandas as pd
import openpyxl
```

```python
#creation of data

#I used AI to produce the following line extracting excel data from Rader_min_plant_data.xlsx
min_plant_data = pd.read_excel('Rader_min_plant_data.xlsx', sheet_name='Sheet1', header=None)

#I used AI to assist in writing the two blocks below, but I debugged the lines. Specifically, I had to tweek the row numbers in the function.

#from the excel file Rader_min_plant_data.xlsx, I extracted rowss 7-19 as x and rows22-34 as y from column G. This is the fractionation data.
x_hendricksite = min_plant_data.iloc[6:19, 6].values #amazonite ε205Tl values
y_hendricksite = min_plant_data.iloc[21:34, 6].values #hendricksite ε205Tl values

#from the excel file Rader_min_plant_data.xlsx, I extracted rowss 7-19 as x and rows22-34 as y from column C. This is the fractionation data.
locations_of_data_x = min_plant_data.iloc[6:19, 2].values #amazonite locations
locations_of_data_y = min_plant_data.iloc[21:34, 2].values #hendricksite locations
```

```python
# production of baisc scatter plot displaying locations of data vs data values

#I used AI to assist in writing the following block.
plt.figure(figsize=(10, 6))
plt.scatter(locations_of_data_x, x_hendricksite, color='blue', label='Amazonite ε205Tl')
plt.scatter(locations_of_data_y, y_hendricksite, color='orange', label='Hendricksite ε205Tl')
plt.xlabel('Location')
plt.ylabel('ε205Tl')
plt.title('ε205Tl Values by Location')
plt.legend()
plt.grid()
plt.show()  
```

```python
#creation of x vs y scatter plot

#I used AI to assist in writing the following block.
plt.figure(figsize=(8, 6))
plt.scatter(x_hendricksite, y_hendricksite, color='green')
plt.xlabel('Amazonite ε205Tl')
plt.ylabel('Hendricksite ε205Tl')
plt.title('Hendricksite vs Amazonite ε205Tl')
plt.grid()
plt.show()
```

```python
#create linear regression

# Ensure x_hendricksite and y_hendricksite are numpy arrays of type float
x_hendricksite = np.array(x_hendricksite, dtype=float)
y_hendricksite = np.array(y_hendricksite, dtype=float)

# Perform linear regression
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x_hendricksite, y_hendricksite)
line = slope * x_hendricksite + intercept
print(f"Slope: {slope}, Intercept: {intercept}, R-squared: {r_value**2}")
# Plot the regression line on the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x_hendricksite, y_hendricksite, color='green', label='Data Points')
plt.plot(x_hendricksite, line, color='red', label='Regression Line')
plt.xlabel('Amazonite ε205Tl')
plt.ylabel('Hendricksite ε205Tl')
plt.title('Hendricksite vs Amazonite ε205Tl with Regression Line')
plt.legend()
plt.grid()
plt.show()
```
